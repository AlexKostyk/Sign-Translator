```python
import numpy as np
import os
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras.callbacks import TensorBoard 
```

```python
model_path = './models/static_model.h5'
```

```python
X_data = []
y_data = []

for root, dirs, files in os.walk('./np_data'):
    for file in files:
        np_arr = np.load(os.path.join(root, file))
        X_data.append(np_arr)
        y_data.append(int(os.path.basename(root)))
```

```python
print(y_data[0])

y_data = np.array(y_data)
X_data = np.array(X_data)

print(X_data.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.10, random_state=99)
```

```python
len(set(y_data))
```

```python
NUM_CLASSES = 68

model = tf.keras.models.Sequential([
    tf.keras.layers.Input((21 * 6, )),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')
])
```

```python
model.summary() 
```

```python
# Model compilation
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
log_dir = os.path.join('mediapipe_hands', 'finale_V2')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
# обучение модели
model.fit(X_train, y_train, epochs=1000, batch_size=64, callbacks=[tb_callback])
```

```python
# Model evaluation
val_loss, val_acc = model.evaluate(X_test, y_test, batch_size=64)
```

```python
# Сохраняем модель
model.save(model_path)
```
